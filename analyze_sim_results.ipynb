```python
from helpers import utils
import pandas as pd
import numpy as np

subfolder_name = "run_20260209_165831"  # Change this to your desired subfolder
sim_data = utils.load_simulation_results(subfolder_name) # Load all results

# Display what was loaded
if sim_data:
    print("\nüìÅ Available data:")
    for filename, data in sim_data.items():
        if isinstance(data, pd.DataFrame):
            print(f"  {filename}: DataFrame with shape {data.shape}")
        elif isinstance(data, np.ndarray):
            print(f"  {filename}: NumPy array with shape {data.shape}")
        elif isinstance(data, dict):
            print(f"  {filename}: Dictionary with {len(data)} keys")
        elif isinstance(data, list):
            print(f"  {filename}: List with {len(data)} items")
        else:
            print(f"  {filename}: {type(data).__name__}")
    
    # Display runtime information if available
    if 'config_readable.json' in sim_data:
        config = sim_data['config_readable.json']
        if 'Runtime' in config:
            print("\n‚è±Ô∏è Simulation Runtime:")
            runtime = config['Runtime']
            print(f"  {runtime['Formatted']}")
    elif 'sim_results_config.csv' in sim_data:
        # Check CSV for runtime info (backward compatibility)
        config_csv = sim_data['sim_results_config.csv']
        if 'Total_Runtime_Seconds' in config_csv.columns:
            runtime_sec = config_csv['Total_Runtime_Seconds'].iloc[0]
            runtime_min = config_csv['Total_Runtime_Minutes'].iloc[0]
            print(f"Simulation Runtime:  {runtime_sec:.2f} seconds ({runtime_min:.2f} minutes)")
        else:
            print("\n‚è±Ô∏è Runtime: Not recorded (older simulation)")
    else:
        print("\n‚è±Ô∏è Runtime: Not recorded (older simulation)")
```
